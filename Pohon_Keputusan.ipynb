3. `Low`: Harga terendah yang dicapai oleh saham selama periode tersebut.
4. `Close`: Nilai harga saham pada penutupan perdagangan pada suatu periode tertentu.

6. `Volume`: Jumlah saham yang diperdagangkan pada periode tersebut.

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import keras
import tensorflow.keras as keras
```

```python
df = pd.read_csv("data_saham.csv")
```

```python
target = df['Open']
print(target)
```

```python
# Menampilkan beberapa baris pertama data
print(df.head())

# Menampilkan beberapa baris terakhir data
print(df.tail())
```

```python
def split_sequence(sequence, n_steps):
  X, y = list(), list()
  for i in range(len(sequence)):
    # Mencari n akhir
    end_ix = i + n_steps
    # Memeriksa urutan terluar dari data
    if end_ix > len(sequence)-1:
      break
    # Menambahkan bagian input dan output
    seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
    X.append(seq_x)
    y.append(seq_y)
  return np.array(X), np.array(y)

# Menentukan Waktu dari data time series
jmlh_X = 3
print('Shapes Target = ',target.shape)
#Split data fitur dan target
X, y = split_sequence(target, jmlh_X)
print('Shapes X = ',X.shape)
print('Shapes Y = ',y.shape)
```

```python
newFitur = pd.DataFrame(X, columns=['Xt-'+str(i) for i in range(jmlh_X-1, -1,-1)])
newTarget = pd.DataFrame(y, columns=['Data Prediksi'])
```

```python
newData = pd.concat([newFitur,newTarget], axis=1)
newData
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
X_norm= scaler.fit_transform(newFitur)
print(X_norm)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=0)
```

```python
# import Decision tree
from sklearn.tree import DecisionTreeRegressor
model_decision_tree = DecisionTreeRegressor()
```

```python
model_decision_tree.fit(X_train, y_train)
y_pred=model_decision_tree.predict(X_test)
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    absolute_percentage_error = np.abs((y_true - y_pred) / y_true) * 100
    mean_error = np.mean(absolute_percentage_error)
    return mean_error

mape = mean_absolute_percentage_error(y_test, y_pred)
print("MAPE:", mape)
```
